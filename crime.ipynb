## Analysis

```python
# Top five most frequent crime types
top_5_crimes = df['Primary Type'].value_counts().head(5)
top_5_crimes
```

```python
#  Is the arrest rate for "ROBBERY" higher or lower than the rate for "NARCOTICS"?
#Filter the DataFrame for the two crime types
robbery_narcotics_df = df[df['Primary Type'].isin(['ROBBERY', 'NARCOTICS'])]

#Group by crime type and calculate total incidents and total arrests
analysis_df = robbery_narcotics_df.groupby('Primary Type').agg(
    total_incidents=('Primary Type', 'size'),
    total_arrests=('Arrest', 'sum')
)

#Calculate the arrest rate
analysis_df['arrest_rate'] = (analysis_df['total_arrests'] / analysis_df['total_incidents']) * 100

analysis_df
```

```python
import matplotlib.pyplot as plt

# Homicide Incidents
homicide_df = df[df['Primary Type'] == 'HOMICIDE']

# Get the total number of homicide incidents
total_homicide = homicide_df.shape[0]

# Get the total number of arrest for homicide
total_homicide_arrest = homicide_df['Arrest'].sum()

# Calculate the arrest rate
arrest_rate_homicide = (total_homicide_arrest / total_homicide) * 100

# Data for the chart
labels = ['Total Homicides', 'Homicides with Arrest']
values = [total_homicide, total_homicide_arrest]

# Create the bar chart
plt.figure(figsize=(8, 6))
plt.bar(labels, values, color=['skyblue', 'salmon'])

# Add labels and a title
plt.title('Homicide Incidents vs. Arrests', fontsize=16)
plt.ylabel('Number of Incidents', fontsize=12)
plt.xticks(rotation=0)

# Add the count on top of each bar for clarity
for i, value in enumerate(values):
    plt.text(i, value + 5, str(value), ha='center', fontsize=11)

# Ensure the y-axis starts from 0 for accurate representation
plt.ylim(0, max(values) * 1.1)

# Save the chart as a file (e.g., as a PNG image)
plt.savefig('homicide_arrest_chart.png')

# Show the plot (optional, as the image file is already saved)
plt.show()
```
